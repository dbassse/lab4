**1. Создание Series из списка**<br>
Создайте Series из списка чисел [5, 15, 25, 35, 45] с индексами ['a', 'b','c', 'd', 'e'] . Выведите его на экран и определите его тип данных.

**2. Получение элемента Series**<br>
Дан Series с индексами ['A', 'B', 'C', 'D', 'E'] и значениями [12, 24, 36,
48, 60] . Используйте .loc[] для получения элемента с индексом 'C' и

```python
s=pd.Series([12, 24, 36, 48, 60],index=['A', 'B', 'C', 'D', 'E'])
print(s.loc['C'])
print(s.iloc[2])
```

**8.Арифметические операции с Series** <br>
Создайте два Series:
- s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
- s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])
  <br>Выполните сложение s1 + s2 . Объясните, почему в результате появляются
NaN , и замените их на 0 .

```python
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])
# Сложение с заменой, NaN появляется из-за того, что индексы в s1 и s2 не одинковые
result = s1.add(s2, fill_value=0)
print(result)
```

**15. Построение графика на основе Series**<br>
Создайте Series , где индексами будут даты с 1 по 30 марта 2024 года, а
значениями – случайные числа от 50 до 150. Постройте график значений с
помощью matplotlib . Добавьте заголовок, подписи осей и сетку.

```python
import matplotlib.pyplot as plt

dates = pd.date_range(start='2024-03-01', periods=30, freq='D')
values = np.random.randint(50, 151, size=30)
ts = pd.Series(values, index=dates)

ts.plot(figsize=(10, 5))
plt.title('График значений')
plt.xlabel('Дата')
plt.ylabel('Значение')
plt.grid(True)
plt.show()
```

**Индивидуальное задание. Динамика спроса на товар в интернет-магазине**<br>
Создайте CSV-файл demand.csv со следующими данными:<br>
Дата,Спрос<br>
2024-09-01,120<br>
2024-09-02,135<br>
2024-09-03,150<br>
2024-09-04,170<br>
2024-09-05,160<br>
Прочитайте файл, установите DatetimeIndex , вычислите скользящее среднее
( rolling(3).mean() ) и постройте два графика: фактического спроса и его
сглаженного значения.

```python
data = '''Дата,Спрос
2024-09-01,120
2024-09-02,135
2024-09-03,150
2024-09-04,170
2024-09-05,160'''
with open('demand.csv', 'w', encoding='utf-8') as f:
    f.write(data)
df = pd.read_csv('demand.csv',parse_dates=['Дата'], dayfirst=False).set_index('Дата')        
df['Скользящее среднее'] = df['Спрос'].rolling(3).mean()
plt.figure(figsize=(10, 7))
plt.plot(df.index, df['Спрос'], marker='o')
plt.plot(df.index, df['Скользящее среднее'], linestyle='--')
plt.title('Динамика спроса на товар')
plt.grid(True)
plt.show()
```
